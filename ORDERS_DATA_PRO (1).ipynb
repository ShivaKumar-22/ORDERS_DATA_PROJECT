# 6. BUSINESS INSIGHTS & RECOMMENDATIONS : 

**PROFIT CALCULATIONS :** 

**Revenue** : (List Price * Quantity) * (1- Discount Percent/ 100)\
**Cost** : Cost Price * Quantity\
**Profit** : Revenue - Cost

```python
# Calcualate Revenue, Cost and Profit

df["Revenue"] = (df["List Price"] * df["Quantity"]) * (1 - df["Discount Percent"] / 100)
df["Cost"] = df["cost price"] * df["Quantity"]
df["Profit"] = df["Revenue"] - df["Cost"]


# Group by Product Id to get the most/least profitable products
product_profit = df.groupby(["Product Id", "Sub Category", "Category"])[["Profit"]].sum().reset_index()

# Most protitable product
most_profitable_product = product_profit.nlargest(1, "Profit")

# least profitablr product
least_profitable_product = product_profit.nsmallest(1, "Profit")

# Gruop by category to get most/least profitable categories
category_profit = df.groupby("Category")[["Profit"]].sum().reset_index()

# Most protitable category
most_profitable_category = category_profit.nlargest(1, "Profit")


# least profitablr category
least_profitable_category = category_profit.nsmallest(1, "Profit")

most_profitable_product, least_profitable_product, most_profitable_category, least_profitable_category
```
